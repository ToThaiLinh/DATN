```python
import pandas as pd
import numpy as np

# Đọc file CSV đầy đủ (chứa các dữ liệu cần cập nhật)
df_full = pd.read_csv('../data/user_event.csv')  # Giả sử là file_full.csv
cols = ['event_time', 'event_type', 'user_id', 'user_session']
df_full = df_full[cols]

# Đọc file CSV chứa product_id và price (file chứa dữ liệu sản phẩm và giá)
df_price = pd.read_csv('../data/product_details_ver3.csv')  # Giả sử là file_price.csv
cols = ['id', 'price']
df_price = df_price[cols]
df_price.rename(columns={
    'id': 'product_id',
}, inplace=True)

# Lấy danh sách product_id từ file_price (có thể có nhiều hơn 1 product_id)
product_ids = df_price['product_id'].tolist()

# Thay đổi cột product_id trong df_full bằng cách chọn ngẫu nhiên từ danh sách product_id
df_full['product_id'] = np.random.choice(product_ids, size=len(df_full))

# Thêm cột price vào df_full bằng cách ghép giá trị price từ df_price theo product_id
df_full = df_full.merge(df_price, on='product_id', how='left')

# Lưu kết quả vào một file CSV mới
df_full.to_csv('../data/user_event_ver2.csv', index=False)
```
