```python
import json
import csv
import os
import time
import random
from kafka import KafkaProducer

# Kafka config
kafka_broker = os.getenv('KAFKA_BROKER', 'kafka1:9092')
topic_name = 'user_event'

producer = KafkaProducer(
    bootstrap_servers=[kafka_broker],
    value_serializer=lambda x: json.dumps(x).encode('utf-8'),
    acks='all',          # đảm bảo broker nhận
    retries=3,           # retry nếu lỗi tạm
    linger_ms=10         # gom message nhẹ
)

csv_file = '../data/user_event_ver2.csv'

with open(csv_file, mode='r', encoding='utf-8') as f:
    reader = csv.DictReader(f)

    for row in reader:
        producer.send(topic_name, value=row)
        # print(f"Sent: {row}")

        time.sleep(0.1)

producer.flush()
producer.close()
```
